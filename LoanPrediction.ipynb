import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Load data explore data

```python
import pandas as pd

# Read the dataset into a DataFrame.
data= pd.read_csv('/content/sample_data/LoanApprovalPrediction.csv')
data.head(5)
obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))

# Dropping Loan_ID column
data.drop(['Loan_ID'],axis=1,inplace=True)


```

convert column into int dataType

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how
# to understand word labels.
label_encoder = preprocessing.LabelEncoder()
obj = (data.dtypes == 'object')
for col in list(obj[obj].index):
  data[col] = label_encoder.fit_transform(data[col])
```

Again check


```python
# To find the number of columns with
# datatype==object
obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))
```

to find any missing value

```python
for col in data.columns:
  data[col] = data[col].fillna(data[col].mean())

data.isna().sum()
```

Spliting DataSet

```python
from sklearn.model_selection import train_test_split

X = data.drop(['Loan_Status'],axis=1)
Y = data['Loan_Status']
X.shape,Y.shape

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
													test_size=0.4,
													random_state=1)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

Define Model

```python
models = [
    Sequential([Dense(7, activation='relu'), Dense(1, activation='sigmoid')]),
    Sequential([Dense(12, activation='tanh'), Dense(1, activation='softplus')]),
    Sequential([Dense(5, activation='leaky_relu'), Dense(1, activation='hard_sigmoid')])
]
```

Compile model

```python

# Compile each model with the same configuration
for model in models:
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

```

Fit Model

```python
i=1
# Train each model
for model in models:
    print("Training %s model :" % (i))
    i += 1
    model.fit(X_train, Y_train, epochs=10, batch_size=32, sample_weight=np.ones(len(Y_train)))
    print("Accuracy score of Model ", id(model), "=", 100 * model.evaluate(X_train, Y_train)[1])
```

Evaluate

```python
i=0
for model in  models:
    _, accuracy = model.evaluate(X_train, Y_train)  # Evaluate each model
    print(f"Accuracy of Model {i+1}: {accuracy*100:.2f}%")
```

Prediction

```python

predictions = []
for model in models:
    model_predictions = (model.predict(X)).astype(int)  # Generate predictions and convert to integers
    predictions.append(model_predictions)
```
